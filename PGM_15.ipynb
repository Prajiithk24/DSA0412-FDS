```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Post_ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Likes': [120, 45, 120, 300, 45, 200, 120, 45]
}
```

```python
df = pd.DataFrame(data)
likes_frequency = df['Likes'].value_counts().sort_index()
```

```python
print("Frequency Distribution of Likes:")
print(likes_frequency)
```

```python
plt.plot(likes_frequency.index, likes_frequency.values)
plt.xlabel("Number of Likes")
plt.ylabel("Number of Posts")
plt.title("Frequency Distribution of Likes Among Posts")
plt.show()
```
