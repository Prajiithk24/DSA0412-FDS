```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

temperature = [18, 20, 25, 30, 35, 38, 36, 34, 32, 28, 23, 19]
rainfall = [30, 40, 35, 50, 60, 120, 150, 140, 90, 60, 45, 35]
```

```python
plt.plot(months, temperature,color='red')
plt.xlabel("Months")
plt.ylabel("Temperature (Â°C)")
plt.title("Monthly Temperature Variation")
plt.show()
```

```python
plt.scatter(months, rainfall,color='green')
plt.xlabel("Months")
plt.ylabel("Rainfall (mm)")
plt.title("Monthly Rainfall Distribution")
plt.show()
```
