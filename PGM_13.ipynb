```python
import string
import matplotlib.pyplot as plt
with open("/content/sample(pgm13).txt", "r") as file:
    text = file.read()
```

```python
text = text.lower()
words = text.split()
word_frequency = {}

for word in words:
    if word in word_frequency:
        word_frequency[word] += 1
    else:
        word_frequency[word] = 1
```

```python

for word, count in word_frequency.items():
    print(word, ":", count)
```
